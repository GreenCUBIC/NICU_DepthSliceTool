```python
import numpy as np
import pandas as pd
from scipy import stats
from matplotlib import pyplot as plt
import itertools
```

```python
results = pd.DataFrame()
results['Tiny_prePT'] = pd.read_csv('ViT_Tiny_prePT.csv').Accuracy.values
results['Tiny_PT'] = pd.read_csv('ViT_Tiny_PT.csv').Accuracy.values
results['Tiny_HHA'] = pd.read_csv('ViT_Tiny_HHA.csv').Accuracy.values
results['Tiny_HHA_PT'] = pd.read_csv('ViT_Tiny_HHA_PT.csv').Accuracy.values
results['Tiny_Sim_prePT'] = pd.read_csv('ViT_Tiny_Sim_prePT.csv').Accuracy.values
results['Tiny_Sim_PT'] = pd.read_csv('ViT_Tiny_Sim_PT.csv').Accuracy.values
results['Tiny_Sim_HHA'] = pd.read_csv('ViT_Tiny_Sim_HHA.csv').Accuracy.values
results['Tiny_Sim_HHA_PT'] = pd.read_csv('ViT_Tiny_Sim_HHA_PT.csv').Accuracy.values
results['Base_prePT'] = pd.read_csv('ViT_Base_prePT.csv').Accuracy.values
results['Base_PT'] = pd.read_csv('ViT_Base_PT.csv').Accuracy.values
results['Base_HHA'] = pd.read_csv('ViT_Base_HHA.csv').Accuracy.values
results['Base_HHA_PT'] = pd.read_csv('ViT_Base_HHA_PT.csv').Accuracy.values
results['Base_Sim_prePT'] = pd.read_csv('ViT_Base_Sim_prePT.csv').Accuracy.values
results['Base_Sim_PT'] = pd.read_csv('ViT_Base_Sim_PT.csv').Accuracy.values
results['Base_Sim_HHA'] = pd.read_csv('ViT_Base_Sim_HHA.csv').Accuracy.values
results['Base_Sim_HHA_PT'] = pd.read_csv('ViT_Base_Sim_HHA_PT.csv').Accuracy.values

results.describe()
```

```python
axes = results.boxplot(figsize=(30, 9))
plt.show()

axes = results.hist(figsize=(15, 15))
plt.show()
```

```python
pair_list = list(itertools.combinations(results.columns, 2))
for (col1, col2) in pair_list:
    value, p1 = stats.normaltest(results[col1].values)
    value, p2 = stats.normaltest(results[col2].values)
    if p1 >= 0.05 and p2 >= 0.05:
        value, pvalue = stats.ttest_ind(results[col1].values, results[col2].values)
    else:
        value, pvalue = stats.ks_2samp(results[col1].values, results[col2].values)
        
    # if pvalue > 0.05:
    #     print(f'{col1} and {col2} difference is not statistically significant')
    # else:
    #     print(f'{col1} and {col2} difference is statistically significant!')

    if pvalue <= 0.05:
        print(f'{col1} and {col2} difference is statistically significant!')
        df = pd.concat([results[col1], results[col2]], axis=1)
        df.boxplot()
        plt.show()
```
